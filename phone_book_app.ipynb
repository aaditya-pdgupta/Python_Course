```python
class Contact:
    phone_directory = [] # It is class variable so it will be share with every object.
    def __init__(self, name, phone_number):
        self.name = name
        self.phone = phone_number
        Contact.phone_directory.append(self)

    def show_contact(self): # It shows the specific contact.
        return f"Name: {self.name}, Contact number: {self.phone}"

    @classmethod # It shows all the conatct.
    def show_all_contact(cls):
        if len(cls.phone_directory) == 0:
            print("No contact found in the directory!!")
        else:
            for contact in cls.phone_directory:
                print(contact.show_contact())
    @classmethod
    def search_contact(cls, search_name):
        for contact in cls.phone_directory:
            if contact.name.lower() == search_name.lower(): # Here is someone mistakely find using lower case than also it will display.
                return contact.phone

        return f"No contact found for {search_name}"

    @staticmethod
    def validate_phone_number(number):
        if len(number) >= 8 and number.isdigit():# It is used to chheck whether all input is digit or not.
            return True 
        else:
            return False
            


    
        
        
    
```

```python
n_contacts = int(input("How many contacts do you want to add?:"))
for i in range(n_contacts):
    name = input("Enter the name of the contact")
    phone_number = input("Enter the phone_number")
    if Contact.validate_phone_number(phone_number):
        Contact(name, phone_number)
    else:
        print(f"Invalid phone number for {name}, phone number should be atleast 8 digits and should only contact numbers")
    
    
```
