#### Morphological Operation
##### Morphological operations are image processing techniques used to modify the shape of objects in an image.
##### Types are:
##### 1. Erosion
-Removes pixels from the boundaries of objects.
##### Syntax 
-          eroded = cv2.erode(img, kernel, iterations=1)

##### 2. Dilations
-Adds pixels to object boundaries.

##### Syntax
-  dilated = cv2.dilate(img, kernel, iterations=1)

##### 3. Opening
-Erosion followed by Dilation

##### Syntax
-opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

##### 4. Closing
-Dilation followed by Erosion.

##### Syntax
- closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

##### 5. Morphological Gradient
##### Morphological Gradient is an operation used to detect the boundary(edge) of objects in an image. It highlights the difference between dilation and erosion of an image.

##### Syntax:
- gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

##### 6. Top-HAt Transformation
##### Top-Hat is a morphological operation used to extract small bright objects from a dark background. It highlights details that are brighter than their surroundings.

##### Syntax:
- cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

##### 7. Black-Hat Transformation
##### Black-Hat is a morphological operation used to extract small dark objects from a bright background. It highlights details that are darker tha their surroundings.

##### Syntax 
- blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

#### Kernel  
##### Kernel (also called a structuring element in morphology) is a small matrix used to process an image.
##### Syntax
- kernel = np.ones((3,3), np.unit8)


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("X_ray.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
h, w = img.shape
```

```python
print("Original Image", w, "x", h)
```

```python
new_width = 256
ratio = new_width / w
new_height = int( h * ratio)
```

```python
resized = cv2.resize(img, (new_width, new_height), interpolation = cv2.INTER_AREA)
```

```python
# If the image is colorful than we need to convert it into the RGB
# resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
```

```python
print("Resized Image", new_width, "x", new_height)
```

```python
kernel = np.ones((5,5), dtype="uint8")
```

```python
erosion = cv2.erode(resized, kernel, iterations=1)
```

```python
dilation = cv2.dilate(resized, kernel, iterations=1)
```

```python
plt.figure(figsize=(8,8))
```

```python
plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original X-ray")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(resized, cmap='gray')
plt.title("Resized X-ray")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(erosion, cmap='gray')
plt.title("Erosion X-ray")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(dilation, cmap='gray')
plt.title("Dilation X-ray")
plt.axis("off")
```
