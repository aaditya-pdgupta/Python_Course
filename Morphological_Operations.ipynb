#### Morphological Operation
##### Morphological operations are image processing techniques used to modify the shape of objects in an image.
##### Types are:
##### 1. Erosion
-Removes pixels from the boundaries of objects.
##### Syntax 
-          eroded = cv2.erode(img, kernel, iterations=1)

##### 2. Dilations
-Adds pixels to object boundaries.

##### Syntax
-  dilated = cv2.dilate(img, kernel, iterations=1)

##### 3. Opening
-Erosion followed by Dilation

##### Syntax
-opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

##### 4. Closing
-Dilation followed by Erosion.

##### Syntax
- closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

##### 5. Morphological Gradient
##### Morphological Gradient is an operation used to detect the boundary(edge) of objects in an image. It highlights the difference between dilation and erosion of an image.

##### Syntax:
- gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

##### 6. Top-HAt Transformation
##### Top-Hat is a morphological operation used to extract small bright objects from a dark background. It highlights details that are brighter than their surroundings.

##### Syntax:
- cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

##### 7. Black-Hat Transformation
##### Black-Hat is a morphological operation used to extract small dark objects from a bright background. It highlights details that are darker tha their surroundings.

##### Syntax 
- blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
