#### Instance Methods
##### Instance methods are functions defined inside a class that work on an objects data. They use the objects instance attributes.

```python
help(list) # They are the instance method
```

```python
class Student():
    """
    This is a class student to manage student info and activities.
    """ 
```

```python
    def Study(): # Here it is class.
        return "The student study for 3 hours per day!"
```

```python
student1 = Student() # The program runs without error.
```

```python
print(student1)  # Here we are getting the memory address.
```

```python
student1.Study()
```
