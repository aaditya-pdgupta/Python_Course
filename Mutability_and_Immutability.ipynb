#### Tuples are immutable.
##### This means we cannot change, modify, add or delete elemnts once the tuple is created.

##### Lists are mutable means they can be chnaged.

```python
s1 = "Python is fun"
```

```python
print(s1.replace("Python", "Java")) # Here it is reassign but not chnaged.
```

```python
print(s1) # The original is same.
```

```python
s2 = s1.replace("Python", "Java")
```

```python
print(s2) # Now we get new string here.
```

```python
t1 =("Mango", "Apple", "Banana")
```

```python
t1.append("Orange) # We can not use the append in tuple.
```

```python
t2 = list(t1)
```

```python
print(t2) # Now here we have changed the tuple in list.
```

```python
t2.append("Orange")
```

```python
print(t2) # Here we are able to add.
```

##### To check the memory address we can use,

```python
print(id(t1)) # Here the memory address is shown.
```
