#### Initializer in Python
##### An initializer is a special method that automatically runs when an object is created. It is used to create and initialize the attributes during the object creation. 

##### Initializer = method that sets initial values of an object. It is an instance method
__ init __


```python
class Student():
    def __init__(self, name, roll, department):
        print(f"Calling the initializer for {self}!!")
        self.name = name
        self.roll = roll
        self.department = department
```

```python
    def study(self, n_hours):
        print(f"The student studies for {n_hours} hours a day!!")
    def sports(self, sports_name):
        print(f"The student plays{sports_name}")
        
```

```python
student1 = Student(name="john", roll=200, department="Biomed") # Here the initializer get called automatically.
student2 = Student(name="Michell", roll=201, department="Civil")
```

```python
print(student1.name, student1.roll, student1.department)
```

```python
print(student2.name, student2.roll, student2.department)
```

```python
print(student1.__dict__)
print(student2.__dict__)
```
