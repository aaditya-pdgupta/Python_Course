#### Recursion
##### Recursion is a programming technique in which a function calls itself to solve a problem by reducing it into smaller sub-problems until a base condition is reached.

##### Factorial of a number n is the product of all positive integers from 1 to n.
##### Notation:
##### n! = n * (n-1) * (n-2) * .....* 1
##### There are two conditions:
##### 1. Base Case---> condition where recursion stops.
##### 2. Recursive case----> function call itself with a smallet input.

```python
# We are not using the recursion.
```

```python
def fact(num):
    factorial = 1
    while num > 1:
        factorial *= num
        num -=1
    return factorial
```

```python
n = 5
```

```python
print(f"Factorial of {n} is {fact(n)}")
```

```python
# Now lets do it using the recusrion
```

```python
def fact_rec(num):
    if num == 1:
        return 1
    else:
        factorial = num * fact_rec(num-1)
        return factorial
```

```python
num = 4
```

```python
result = fact_rec(num)
```

```python
print(f"Factorial of {num} is {result}")
```
