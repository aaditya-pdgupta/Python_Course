#### filter() with lambda
##### Select elements that statisfy a condition
##### Syntax:
                filter(lambda arguments: condition, iterable)

```python
# filter(function, seq)
```

```python
seq = [1, 2, 4, 5]
```

```python
# Here we want to keep the odd number and remove the even number.
```

```python
odd = lambda x: True if x % 2 != 0 else False
```

```python
filtered_out = filter(odd, seq)
```

```python
print(filtered_out) # Here we are getting the memory location.
```

```python
#print(list(filtered_out)) # When we are using the list we are getting the proper output.
```

```python
print(f"Odd number in the above sequence are: {list(filtered_out)}")
```

#### map() with lambda
##### Takes each value in a list and chnages it using a rule.
##### Syntax:
                map(lambda arguments: condition, iterable)

```python
s = [1, 2, 3, 4, 5]
```

```python
odd = lambda x: True if x % 2 != 0 else False
```

```python
filtered_output = map(odd, s)
```

```python
print(f"Map output: {list(filtered_output)}") # It gives the condition. Map is not suitable for this. It is useful for opertaing certain operation.
```

```python
square = lambda x: x ** 2
```

```python
filtered= map(square, s)
```

```python
print(f"Square number : {list(filtered)}") 
```
