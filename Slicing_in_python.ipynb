### Slicing in the Python
Slicing means extraxting a part (portion) of a sequence like string, list or tuple.

 syntax:
         sequence[start : stop : step]

         Meaning
             Start: index where slicing begins (included).
             stop: index where slicing ends (excluded)
             step: jump size (default = 1)


```python
#syntax of idexing : string[index]
#syntax of slicing : string[start:end:step]
```

```python
print(s1[2:7:1]) # Here if we count the Hello World than H=0, e=1, l=2, ......and d= 10. here the space is also counted. So here 2 indicates the start so it will take l as first and stop at 7 means o so it is excluded and step is 1 so it will jump one only.
```

```python
print(s1[2:9:2]) # Here it start from the 2 and ends at 9 having jump 2.
```
