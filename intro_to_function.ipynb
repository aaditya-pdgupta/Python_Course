#### Function
##### A function is a block of reusable code that performs a specific task.
##### Instead of writing the same code again and again, we define it once and call it whenever needed.
##### Syntax
            def function_name(parameters):
                statements 
                return value
            
                

##### Built-in Function (Python)

##### Definition:
##### A built-in function is a function that is already provided by Python, so you don’t need to define or import it. You can use it directly.

##### Why built-in functions are used?

##### Save time ⏱️

##### Reduce code length

##### Already tested and optimized

#### Common Built-in Functions (Exam-friendly)

##### print() ---	Display output---	print("Hello")
##### input()----	Take user input	---x = input()
##### len()	---- Length of string/list---- 	len([1,2,3])
##### type()----	Check data type----	type(10)
##### int()	----- Convert to integer --- 	int("5")
##### float() -----	Convert to float	---- float("3.2")
##### str()	---- Convert to string ------str(100)
##### sum() ---- Sum of elements ---- sum([1,2,3])
##### max()	---- Maximum value ----	max(4,9,2)
##### min()	---- Minimum value ----	min(4,9,2)
##### range() ----	Generate sequence ----- range(1,5)
##### abs()	---- Absolute value ----	abs(-10)
###### round() ----	Round number ----	round(3.6)
