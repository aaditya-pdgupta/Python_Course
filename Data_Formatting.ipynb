#### Data Formatting 
##### Data Formatting is the process of changing the apperance, structure or data type of data without changing its meaning, so it becomes consistent, readable and already for analysis.

```python
import pandas as pd
df = pd.read_excel("Student_info.xlsx")
```

```python
print(df)
```

```python
df["Marks"].head()
df["Marks"].max() # Here we are comfirming either it is in the percentage or not. So the output is greater than the 1 means it is not in percentage.
```

```python
df["Marks"] = df["Marks"] / 5 # Here we are dividing the entire column by the 5.
```

```python
df.head()
```

```python
df.rename(columns = {"Marks":"Obtained_Marks"}, inplace = True) # Here we have renamed the marks as the Obtained_Marks.
```
