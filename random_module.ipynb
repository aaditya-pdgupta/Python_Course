#### random module in Python 
###### The random module is used to generate random numbers and choices.

#### 1. random.random()
##### Returns a random float between 0.0 and 1.0 (excluded).

import random # Returns random float between 0.0 to 1.0 (excluded).
print(random.random())
print(random.random())
print(random.random()) # Each and every time of execution it gives the different value.



#### 2. random.randint(a, b)
##### Returns a random integer between a and b (both included).

```python
print(random.randint(1, 10)) # Each time of execution it gives the random value.
print(random.randint(1, 10))
```

#### 3. random.randrange(start, stop, step)
##### Like range() but returns one random value.

```python
print(random.randrange(0, 20, 2)) #Each time of execution it gives the random value.
print(random.randrange(0, 20, 2))

```

#### 4. random.choice(sequence)
##### Select one random element from a list/tuple/string.

```python
nums = [10, 4, 1, 20, 5]
```

```python
print(random.choice(nums))  #Each time of execution it gives the random items from the sequence.
print(random.choice(nums))
```

```python
fruits = ["Apple", "Banana", "Mango"]
```

```python
print(random.choice(fruits)) #Each time of execution it gives the random items from the sequence.
print(random.choice(fruits))
```

#### 5. random.shuffle(list)
##### Shuffles a list in place.

```python
random.shuffle(fruits) # Here it shuffle randomly the list. It can be used in the playing card.
print(fruits)
```
