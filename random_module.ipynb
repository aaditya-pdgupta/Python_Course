#### random module in Python 
###### The random module is used to generate random numbers and choices.

#### 1. random.random()
##### Returns a random float between 0.0 and 1.0 (excluded).

import random # Returns random float between 0.0 to 1.0 (excluded).
print(random.random())
print(random.random())
print(random.random()) # Each and every time of execution it gives the different value.



#### 2. random.randint(a, b)
##### Returns a random integer between a and b (both included).

```python
print(random.randint(1, 10)) # Each time of execution it gives the different value.
print(random.randint(1, 10))
```

#### 3. random.randrange(start, stop, step)
##### Like range() but returns one random value.

```python
print(random.randrange(0, 20, 2)) #Each time of execution it gives the different value.
print(random.randrange(0, 20, 2))

```
