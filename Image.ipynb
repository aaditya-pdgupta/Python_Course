#### What is Image?
##### An image is a collection of pixels arranged in rows and columns.

#### Types of Images
##### 1. Binary Images (Balck and White)
- Only 2 values: 0 or 255.
- -Black (0) or Whites(1).
- Obtained using thresholding.

##### 2. Grayscale Image 
- Single channel
- -Value from 0-255
- -Different shades of gray

##### 3. Color Images (RGB/BGR)
- 3 channels
- Red, Green, Blue
- OpenCv uses BGR

##### 4. Multichannel/ Multispectral Image 
-More than 3 channels.
-Each channel represents different wavelength.

##### 5.3D Image (Volumetric Image)
-Stack of 2D slice.
-Forms 3D volume.

##### Image Format
##### Image format = how and image file is stored on disk.
- JPEG/JPG (Joint PHotographic Experts Group)
- PNG (Portable Network Graphics)
- BMP (Bitmap)
- TIFF (Tagged Image File Format)
