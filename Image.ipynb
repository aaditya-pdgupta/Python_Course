#### What is Image?
##### An image is a collection of pixels arranged in rows and columns.

#### Types of Images
##### 1. Binary Images (Balck and White)
- Only 2 values: 0 or 255.
- -Black (0) or Whites(1).
- Obtained using thresholding.

##### 2. Grayscale Image 
- Single channel
- -Value from 0-255
- -Different shades of gray

##### 3. Color Images (RGB/BGR)
- 3 channels
- Red, Green, Blue
- OpenCv uses BGR

##### 4. Multichannel/ Multispectral Image 
-More than 3 channels.
-Each channel represents different wavelength.

##### 5.3D Image (Volumetric Image)
-Stack of 2D slice.
-Forms 3D volume.

##### Image Format
##### Image format = how and image file is stored on disk.
- JPEG/JPG (Joint PHotographic Experts Group)
- PNG (Portable Network Graphics)
- BMP (Bitmap)
- TIFF (Tagged Image File Format)
- DICOM (Digital Imaging and Communication in medicine)

##### Applications
- Medical diagnosis (MRI, CT, X-ray)
- Tumor and disease detection
- Fracture identification
- Face recognition systems
- CCTV surveillance
- Self-driving cars
- Industrial defect detection
- Robot vision systems
- Satellite imaging
- Mobile camera filters & face unlock


#### Computer Vision (OpenCV)
##### Computer Vision is a field of Artificial Intelligence that enables computer to see, understand and analyze images and videos. It detect objects, Recognize faces, Identify patterns, Analyze medical scans, Track movement etc.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("CVV.jpg") # Here it read the image.
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)# OpenCv reads image in BGR but matplotlib displays image in RGB so it conert in required format.
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
