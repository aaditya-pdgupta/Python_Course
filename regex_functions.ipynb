##### match()
##### Checks only at the start of the string.

```python
import re
```

```python
s1 = "We are learning regex in Python"
```

pat = r"[A-Z][a-z]"

```python
match = re.match(pat, s1)
```

```python
print(match)
```

```python
# Let us see are from the s1 than.
```

```python
pat = r"[a-z]{3}"
```

```python
match = re.match(pat, s1)
```

```python
print(match)  # Here we are not getting because match only look for the starting only.
```

#### Search 
##### Seraches anywhere in the string.

```python
phones = "John-1234567890, Carol-9855442211, Mark- 8877555665"
```

```python
pat = r"[0-9]{10}"
```

```python
match = re.search(pat, phones)
```

```python
print(match)
```

##### findall
##### Returns all matches are as list.

```python
pat = r"[0-9]{10}"
```

```python
match = re.findall(pat, phones)
```

```python
print(match) # Here we are getting all the possible matches.
```

```python
# Let us see if all the number are not of 8 digit than
```

```python
phn = "John-1234567890, Carol-9855442211, Mark- 8877555665, Alice-222222, Python 3.14" 
```

```python
pat = r"[0-9]{10}"
```

```python
match = re.findall(pat, phones)
```

```python
print(match) # Here we are getting only the numbers containg 10 digit.
```

```python
pat = r"[0-9]+"
```
