##### match()
##### Checks only at the start of the string.

```python
import re
```

```python
s1 = "We are learning regex in Python"
```

```python
pat = r"[A-Z][a-z]"
```

```python
match = re.match(pat, s1)
```

```python
print(match)
```

```python
# Let us see are from the s1 than.
```

```python
pat = r"[a-z]{3}"
```

```python
match = re.match(pat, s1)
```

```python
print(match)  # Here we are not getting because match only look for the starting only.
```
