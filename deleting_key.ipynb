```python
# We have following dictionary containg the detail. We need to delete the sensitive key such as password addresss from that.
```

```python
user = {"user_name": "Aadi", "password": "admin", "email": "abc@123.com", "address": "Vadodara", "country": "India"}
```

```python
sensitive_info = ["password", "address"]
```

```python
for i in sensitive_info: #If we are running the loop in dict or list and trying to change same dict or list it will give the run time error. Run it on the data type which need to be modify.
    user.pop(i)
    
print(user)

```

```python
sensitive_info = ["password", "address", "Phone"]
```

```python
# In above we are getting the error. The way is given below.
```

```python
for i in sensitive_info:
    if i in user:
        print(f"Deleted ==> key: {i}, value: {user[i]}")
        user.pop(i)
    else:
        print(f"{i } not present")
  
    
print(user)
```
