#### Copy Library in Python
##### The copy module in Python is used to create shallow copies and deep copies of objecs. This is important because normal assignment does not copy objects, it only copies references.

```python
import copy
```

```python
l1 = [1, 2.2, [10, 20, 30], "Python"]
```

```python
type(l1)
```

```python
len(l1)
```

#### Shallow copy
##### A shallow copy created a new object, but the inner/nested objects are NOT copied. 
##### Syntax: 
                copy.copy()
