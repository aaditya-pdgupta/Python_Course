#### Copy Library in Python
##### The copy module in Python is used to create shallow copies and deep copies of objecs. This is important because normal assignment does not copy objects, it only copies references.

```python
import copy
```

```python
l1 = [1, 2.2, [10, 20, 30], "Python"]
```

```python
type(l1)
```

```python
len(l1)
```

#### Shallow copy
##### A shallow copy created a new object, but the inner/nested objects are NOT copied. 
##### Syntax: 
                copy.copy()

```python
l2 = copy.copy(l1) # Here the value is same for the l1 and l2 but their memory adderess is different.
```

```python
print(l2)
```

```python
print(id(l1)) #The memory addresses for the both l1 and l2 is different.
```

```python
print(id(l2))
```

```python
l1[0] =  100 # We have changed the value of the index 0 in l1.
```

```python
print(f"This is value of l1 {l1}")
```

```python
print(f"This is the value of l2 {l2}")  # But the value in l2 remain same of before. NO value has been updated although the value of l1 has been updated.
```

```python
l1[2][0] = 5
```

```python
print (l1)
```

```python
print(l2) 
```

##### Above when we changed the element of the l1 directly there was no impact on l2. The value remain unchnaged in the l2. But when we changed the value of inner list that is list inside list the l1 and l2 both get updated. So simply for list or outer object it will not chnage but for the nested list it will get updated

#### Deep Copy
##### A deep copy creates a completely independent copy of an object, including all nested objects inside it. So changes in the copied do NOt affect the original.
##### Syntax:
            copy.deepcopy()

```python
l3 =[4, 5.5, [8, 9, 10], "Java"]
```

```python
l4 = copy.deepcopy(l3)
```

```python
print(l4)
```

```python
#l3[1] = 0 
```

```python
print(l3)
```

```python
l3 [2][2] = 0 # Here we have chnaged the nested loop value.
```

```python
print(l3) # Here the value of l3 get changed.
```

```python
print(l4) # But the value of l4 is still previous one. So it clarify there is no chnage in the nested loop using the deep copy.
```

##### In simply we can say using deep there is no change in either list or nested list but using the copy there wont be chnage in the main list but the nested list  get chnaged.
