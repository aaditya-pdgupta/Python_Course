#### Compile ()
##### re.compile is used to compile a regular expression into a pattern object, which can be reused multipile times.
##### Syntax:
##### re.compile(pattern, flags=0)


```python
import re
```

```python
phones = "Alice-123456254, Mark-7887887888, Carol-123578887"
```

```python
pattern = r"\d{10}"
```

```python
match_obj = re.findall(pattern, phones)
```

```python
print(match_obj)
```
