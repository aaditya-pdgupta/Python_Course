#### Inheritance 
##### Inheritance means one class(child) acquires the properties and methods of another class (parent).
##### Syntax:
            class Parent:
              #Parent code

              class Child(Parent)
                  #child code

```python
class Vehicle:
    company = "XYZ Motors"

    def __init__(self, n_wheels, n_seats, mileage):
        self.n_wheels = n_wheels
        self.n_seats = n_seats
        self.mileage = mileage

    def get_details(self):
        return f"This vehicle has {self.n_wheels}, {self.n_seats} seats and provide a mileage of {self.mileage}."
```

```python
v1 = Vehicle(n_wheels = 4, n_seats = 7, mileage = 70)#
```

```python
print(v1.get_details())#
```

```python
class Car(Vehicle): 
    model = "ABC12"
    def __init__(self, car_type, drive_type):
        print("init of Car")
        self.car_type = car_type
        self.drive = drive_type
        Vehicle.__init__(self, n_wheels = 4, n_seats = 7, mileage = 15)
```

```python
# Car Class inherits the Vehicle class.
# Car Class is called as Child class/derived class.
#Vehicle class is called Parent class/base class.
```

```python
c1 = Car() # Here we are getting the error in child class because there is no init of child class so it uses the parent class init.
```

```python
c1 = Car(car_type = "SUV", drive_type = "Manual") 
```

```python
print(c1.model)
```

```python
print(c1.company)
```

```python
print(c1.mileage) # Here we are getting the error because the child class has its own init. So it call its own init.
```

```python
print(c1.get_details())
```
