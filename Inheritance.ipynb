#### Inheritance 
##### Inheritance means one class(child) acquires the properties and methods of another class (parent).
##### Syntax:
            class Parent:
              #Parent code

              class Child(Parent)
                  #child code

```python
class Vehicle:
    company = "XYZ Motors"

    def __init__(self, n_wheels, n_seats, mileage):
        self.n_wheels = n_wheels
        self.n_seats = n_seats
        self.mileage = mileage

    def get_details(self):
        return f"This vehicle has {self.n_wheels}, {self.n_seats} seats and provide a mileage of {self.mileage}."
```

```python
v1 = Vehicle(n_wheels = 4, n_seats = 7, mileage = 70)#
```

```python
print(v1.get_details())#
```

```python
class Car(Vehicle): 
    pass
```

```python
# Car Class inherits the Vehicle class.
# Car Class is called as Child class/derived class.
#Vehicle class is called Parent class/base class.
```
