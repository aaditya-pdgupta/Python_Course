#### Regular Expression (RegEx)
##### A regular expression is a sequence of characters that defines a search pattern.

```python
import re # HEre we are importing the Regex Module 
```

```python
message = "The current Python version is 3.15. Other previous versons are 3.14, 3.13 and so on."
```

```python
# If we want to check wether python is present in the message or not tha.
```

```python
print("Python" in message)  # Here it check whether the Python is present in the message or no
```

```python
print("15" in message) #It is giving as true.
 
```

```python
print("14" in message) # Here it is giving false because there is no 14 in the message.
```

```python
print(message.find("3.15")) # Here we are getting the 30 which is the index of 3.15.
```

```python
print(message.find("Python")) # Here we are getting the index of the P .
```

```python
# We will use Regex module.
```

```python
#re.search(regex_pattern, string) => returns a match object when there is a match found, else returns NOne.
```

#### re.search()
##### Search anywhere in string.

```python
match_obj = re.search("13", message)
```

```python
print(match_obj) 
```

```python
# The explanation of the above output is :
#  span = (71, 73)
#        ==> The match starts at index 71.
#        ==> The match ends at index 73.
#  match='13'
#         ==> The actual text that matched the pattern.
```

```python
if match_obj:
     print("found")
else:
    print("Not found")

print(match_obj)
    
```

```python
if re.search("13", message):
    print("found")
else:
    print("Not found")
```
