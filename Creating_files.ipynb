#### Creating file
##### A file is created using the open() function.
##### Syntax:
#####        open("filename", "mode")


#### File Creation Modes
##### 1. "x" Mode (Create only)
##### 2. "w" Mode (Create + Write)

```python
file = open("file1.txt", "x") # Here we have created a file
```

```python
# Writing into a file
# write(content)
```

```python
file.write("This file is created using the 'x' mode in Pytho.\n")
```

```python
file.write("Next line.")
```

```python
# Closing the file
```

```python
file.close()
```
