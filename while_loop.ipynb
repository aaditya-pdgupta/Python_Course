#### While loop
#####1. Repeats a block of code as long as a condition is True.
#####2. Used when number of iterations is not fixed.
##### Syntax:
            while condition:
                statements

```python
for i in range(1, 5, 1):
    print(i)
```

```python
num = 1 
```

```python
while num < 5: # Here the iteration continues til
    print(num)
    num = num + 1
```

```python
while True: # Infinite loop.
    user_password =  input("Enter your password:") # Here user enter his/her password
    if user_password == correct_password: # It is the condition if password of user is matched with correct password than it will print and break the loop.
        print("Password is correct.")
        break
print("Logged in!!!")
```

```python
number_of_times = 0
correct_password = 12345
max_attempt = 3
logged_in = False
```

```python
while number_of_times < max_attempt:
    user_password =  input("Enter your password:") 
    if user_password == correct_password: 
        print("Password is correct.")
        logged_in = True
        break
    else:
        number_of_times += 1
        print("You havve entered wrong password, try again.")
if logged_in:
    print("Congrats!!!, Logged in successfully..")
else:
    print("Your account has been blocked! Too many wrong attempts.")

    
```
