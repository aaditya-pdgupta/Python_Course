#### While loop
#####1. Repeats a block of code as long as a condition is True.
#####2. Used when number of iterations is not fixed.
##### Syntax:
            while condition:
                statements

```python
for i in range(1, 5, 1):
    print(i)
```

```python
num = 1 
```

```python
while num < 5: # Here the iteration continues til
    print(num)
    num = num + 1
```

```python
correct_password = "Python"
```

```python
while True: # Infinite loop.
    user_password =  input("Enter your password:") # Here user enter his/her password
    if user_password == correct_password: # It is the condition if password of user is matched with correct password than it will print and break the loop.
        print("Password is correct.")
        break
print("Logged in!!!")
```

```python
# Here we have seen the user entering trhe wrong password is not accepted and its keep on asking. Lets print something fro the wrong password.
```
