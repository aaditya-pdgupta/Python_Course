#### While loop
#####1. Repeats a block of code as long as a condition is True.
#####2. Used when number of iterations is not fixed.
##### Syntax:
            while condition:
                statements

```python
for i in range(1, 5, 1):
    print(i)
```

```python
num = 1 
```

```python
while num < 5: # Here the iteration continues til
    print(num)
    num = num + 1
```
