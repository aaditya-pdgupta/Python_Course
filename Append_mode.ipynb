#### Append Mode 
##### Append means adding new data at the end of an existing file without deleting old data.

##### Syntax
#####               open("filename.txt", "a")

```python
file = open("Practice.txt", "a")
```

```python
file.write("\nThis has content has been written using the 'a' mode.'n")
file.write("'a' mode is used to add new content at the end of the file.\n")
file.write("Good bye!")
```

```python
file.close()
```

```python
print(file)
```
