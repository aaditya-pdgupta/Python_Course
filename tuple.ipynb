#### Tuple
##### A tuple is an ordered, immutable (cannot be chnaged) collection of elements. A tuple is written using parentheses(), not square brackets.
#### Syntax:
    (items1, items2, items3)

##### Properties of Tuple
1. ##### Ordered---Elements have fixed positions (indexing allowed).
2. ##### Immutable---Cannot change elements after creation.
3. ##### Allows duplicates----Same value can appear multiple times.
4. ##### Can store mixed data types(int, float, string, list, another tuple etc).

##### Tuples are used when the data should be constant like days of week, coordinates (x,y,z), students rollnumber, DOB, Database record).

```python
t1 = ("Python", 10, 1.5, True, [1, 2, 3], (10, 20)) # Here we can store anything.
```

```python
print(t1)
```

```python
len(t1) # here we can use the len function to find the length of tuple.
```
