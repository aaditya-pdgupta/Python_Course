#### Class Method
##### A class method is a method that is bound to the class, not to an object. It can access and modify class variables. It uses cls as the first parameter and is defined using the @classmethod decorator.

          entsts



##### Class Method Syntax

```python
class ClassName:
    @classmethod
    def method_name(cls, arguments):
        statements

````python
class Welcome:
    @classmethod
    def greet(cls): # here we have class method greet. 
        print("HEllo")
````

````python
w1 = Welcome()
````

````python
w1.greet()
````

````python
print(Welcome)
````

````python
class Student:
    college_name = "ABC College" 
    departments = ["Arts", "Commerce", "Science"]

    def __init__(self, name, roll): 
        self.name = name
        self.roll = roll

    def study(self, n_hours):
        print(self)
        print(f"The student studies for {n_hours} hours a day ! ")
        
    def sports(self, sports_name):
        print(f"The student plays {self.sports_name}")

    @classmethod
    def greeting(cls):
        print(cls)
        print(f"Welcome to the college {cls.college_name}!!") # Here the cls use for the class name and it fetch the class variable.
        
    
````

````python
student1 = Student(name = "John", roll = "1111")
````

````python
student1.study(3)
````

````python
# Now we need to call the class method.
````

````python
student1.greeting()
````

````python
student1.sports("Football")
````
