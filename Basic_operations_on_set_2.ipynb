```python
student1 = {"English", "Math", "CS", "Physics", "Chemistry"}
```

```python
student2 = {"English", "Biology", "CS", "Physics", "Chemistry"}
```

```python
print(student1, type(student1)) # We have printed the student 1 detail as well as type to comfirm it.
```

```python
# Common subject of student 1 and student 2. - intersection
```

#### Intersection
##### Intersection means finding the common elements between two sets (the values that appear in both).
##### Syntax:
                first_element.intersection(second_element)

```python
common_subjects = student1.intersection(student2)
```

```python
print(common_subjects) # We can print by storing in the variable .
```

```python
print(student1.intersection(student2)) # We can direct print also.
```

```python
print(student1 & student2) # We can get the intersection using the & also.
```

#### Union
##### Union means combining all elements from two sets, but without duplicates.
##### Syntax:
                a.union(b)

```python
print(student1.union(student2)) # Here we have common subject without repetition of the duplicate one.
```

```python
student3 = {"Sanskrit", "Maths", "CS"} # Lets add one more student and find the intersection and union.
```

```python
print(student1.intersection(student2, student3)) # So here the one common subject is there between all the three student.
```

```python
print(student1.union(student2, student3))  #  Here all the subject without repetition between three are displayed below.
```
