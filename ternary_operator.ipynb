#### Short-hand if/ Ternary Operator in Python
##### The ternary operator is one line shortcut for writing simple if-else statements.
##### Syntax:
            value_if_true if condition else value_if_false

```python
num = int(input("Enter a number:"))
```

```python
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
 
```
