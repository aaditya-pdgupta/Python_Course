#### Passing function as arguments
##### Passing a function as an argument mean you can give one function to another function and the second function can all/use it.


```python
def add_1(num): # So it is the simple function.
    return num +1
```

```python
print(add_1(10)) # It gives the output as 11.
```

def square(num):
    return num** 2

```python
print(square(4))
```

```python
num = int(input("Enter a number:"))
```

```python
res_1 = add_1(num)
```

```python
res_2 = square(res_1)
```

```python
print(f"Result is {res_2}")
```

```python
#Instead Using the two function we can do
```

```python
res = square(add_1(5))
```

```python
print(f"Result is {res}")
```
