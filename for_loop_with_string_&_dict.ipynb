#### For Loop with String
##### A string is a sequence of characters, so the for loop iterates character by character.

```python
s1 = "Hello World"
```

```python
for char in s1: # Here we are able to print each character of the sentence.
    print(char)
print("End of the loop") # It is outside of the loop which will be printed once the loop get ended.
```

#### For Loop with Dictionary
##### A dictionary contains key-value pairs.

```python
employee = {"id": 1010, "name": "John Gray", "Department": "Biomedical"}
```

```python
for i in employee: # It will print the keys only.
    print(i)
```

```python
for i in employee: # Here when we print i the key get printed and when employee {i} here the vlaue also get printed.  
    print(i, employee[i])
```

```python
print(employee.items()) # Here we are getting the key and value in the form of tuple.
```

```python
for i in employee.items():  # here i has the key value pair so it present each.
    print(i)
```

```python
for i in employee.items():  # Here the o represent the key.
    print(i[0])
```

```python
for i in employee.items(): 
    print(i)
```
