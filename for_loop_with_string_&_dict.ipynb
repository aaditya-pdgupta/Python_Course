#### For Loop with String
##### A string is a sequence of characters, so the for loop iterates character by character.

```python
s1 = "Hello World"
```

```python
for char in s1: # Here we are able to print each character of the sentence.
    print(char)
print("End of the loop") # It is outside of the loop which will be printed once the loop get ended.
```

#### For Loop with Dictionary
##### A dictionary contains key-value pairs.

```python
employee = {"id": 1010, "name": "John Gray", "Department": "Biomedical"}
```

```python
for i in employee: # I
    print(i)
```
