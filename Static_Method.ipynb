#### Static Method
##### A static metod is a method that belongs to a class but does not use instance (self) or class (cls) data. It behaves like a normal function, but is placed inside a class for logical grouping.

nts

##### Syntax

                    class ClassName:
                        @staticmethod
                        def method_name(arguments):
                            statements

```python
class Student:
    college_name = "ABC College" 
    departments = ["Arts", "Commerce", "Science"]

    def __init__(self, name, roll): 
        self.name = name
        self.roll = roll

    def study(self, n_hours):
        print(self)
        print(f"The student studies for {n_hours} hours a day ! ")
        
    def sports(self, sports_name):
        print(f"The student plays {sports_name} in the college {self.college_name}")

    @staticmethod
    def greeting():
        print(cls)
        print(f"Welcome to the college {cls.college_name}!!") # Here the cls use for the class name and it fetch the class variable.

    @classmethod
    def get_department(cls):
        print(f"Departments in {cls.college_name} are:")
        for department in cls.departments:
            print(department)
            
        

    
```
