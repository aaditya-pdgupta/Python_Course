#### Nested Loop
##### A nested loop means a loop inside another loop. The inner loop runs completely for each iteration of the outer loop.

##### Syntax:
            for i in range(outer_limit):
                for j in range(inner_range):
                         statement

```python
for i in range(3): # It is outer loop.
    for j in range(2): # It is inner loop
        print(f"i={i}, j={j}")
    
```

```python
for i in range(3): # It is outer loop.
    for j in range(5): # It is inner loop
        print(f"i={i}, j={j}")
    
```
