#### Flipping Image
##### Flipping means mirror the image.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("highway.jpg")
```

```python
h,w = img.shape[:2]
print("Original Image", w, "x", h)
```

```python
new_width = 256
ratio = new_width / w
new_height = int(h * ratio)
```

```python
resize_img = cv2.resize(img, (new_width, new_height), interpolation = cv2.INTER_AREA)
```

```python
resized_rgb = cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(resized_rgb)
plt.title("Original Image")
plt.axis("off")
```

```python
flip = cv2.flip(resized_rgb,1) # Here 1 means flipping horizontally.
```

```python
plt.imshow(flip)
plt.title("Flipped Image")
plt.axis("off")
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(resized_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(flip)
plt.title("Horizontal Flipped Image")
plt.axis("off")

```

```python
flip_1 = cv2.flip(resized_rgb,0) # Here using the 0 means flipping vertically.
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(resized_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(flip_1)
plt.title(" Vertical Flipped Image")
plt.axis("off")
```
