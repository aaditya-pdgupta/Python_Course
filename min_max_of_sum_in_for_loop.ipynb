```python
scores = [2, 45, 102, 4, 9, 12, 45, 90, 1, 0, 1]
```

```python
print(len(scores))
```

```python
# Suppose we have to find the sum of the scores.
```

```python
for score in scores: # Doing this print all the score turn by turn.
    print(score)
```

```python
total = 0
```

```python
for score in scores: 
    total += score
print(f"The total score is {total}")
```

```python
# Without using the for loop we can use below sum function directly.
```

```python
total = sum(scores)
```

```python
print(f"The total score is {total}")
```
