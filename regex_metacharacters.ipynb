#### What are the Metacharacters?
##### metacharacters are special characters that have a special meaning in regular expression and are used to define patterns.

```python
import re
```

```python
message = "The current Python version is 3.15. Other previous versons are 3.14, 3.13 and so on."
```

```python
match_object = re.search("[0-9][0-9]", message)
```

```python
print(match_object)
```

##### Lets see why 15 is match not other.
##### first it check 3. not match becaue . is not digit.
##### than it check .1 is not match because . is not a digit.
##### Than it check 15 which is perfect match and stop there and do not check other further.

```python
match = re.search("[0-9][0-9]", "House number: 25")
```

```python
print(match)
```

#### . dot
##### The dot (.) is a metacharacter that matches any single character
✔ letters
✔ digits
✔ spaces
✔ symbols

❌ except newline (\n) (by default)

```python
match_obj = re.search("[0-9].[0-9][0-9]", message)
```

```python
print(match_obj)
```

```python
match_obj = re.search("[0-9].[0-9]", message)
```

```python
print(match_obj)
```

```python
mess = "The year is 2026"
```

```python
matc = re.search("[0-9].[0-9][0-9]", mess)
```

```python
print(matc)
```

#### Explanation. 
##### 1. [0-9] matches with 2.
##### 2. . dot matches any character
##### 3. [0-9] matches 2 and [0-9] matches 6.

```python
# Let us see we have to match the dot as particular condition So we will keep in in square bracket. [.] 
```

```python
matc = re.search("[0-9][.][0-9][0-9]", message)
```

```python
print(matc)
```
