#### Polynomial Regression
##### Polynomial Regression is an extension of Linear Regression where we fit a curve instead of a straight line.

#### Curvilinear Relationship 
##### A curvilinear relationship is non linear relationship where the chnage between two variables follows a curved pattern instead of straight line.

```python
import numpy as np
x = np.random.randn(100)
y =np.random.randn(100)
```

```python
f = np.polyfit(x, y, 3) # Fit a polynomia of degree 3 (cubic curve) to data (x,y). We can change the order.
p = np.poly1d(f) # This converts the coefficient into a polynomial function.
```

```python
print(p)
```

#### Sample Evaluation
##### Sample Evaluation is the process of measuring model  performance using a subset of data (test sample) that was not used duringg training.

##### Why does we use It ?
- To check model performance on unseen data.
- To detect the overfitting.
- To measure real- world prediction accuraccy.
- To measure model generalization

##### Dta Splitting
##### Dataset is divided into:
- Training Sample(70-80%) ---> Used to train model.
- Testing Sample (20-30%) ---> Used for  evaluation.

##### Commonn Evaluation Metrices (Regression)
- $R^2$ Score -----> Goodness of fit.
- Mean of Squared Errors (MSE) ----> Average squared error.

#### Mean Squared Error (MSE)
##### Mean Squared Error (MSE) is a regression evaluation metric that measures the average of the squared diffrences between actual and predicted values.

```python
from sklearn.metrics import mean_squared_error # It imports the Mean Squared Error (MSE) function from the sklearn.metrics module. 
flights = sns.load_dataset("flights")
```

```python
mean_squared_error(flights['year'], flights['passengers'], multioutput = 'raw_values')
```

```python
mean_squared_error(flights['year'], flights['passengers'], multioutput = 'uniform_average')
```

```python
mean_squared_error(flights['year'], flights['passengers'], squared = True)
```

```python
mean_squared_error(flights['year'], flights['passengers'], squared = False) #It gives the root MSE.
```
