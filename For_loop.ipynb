#### For loop 
##### A for loop is used to repeat a block of code for each item in a sequence like:
##### ---> list
##### ----> tuple
##### ----> string
##### -----> range
##### ------> set
##### -------> dictionary
##### It is also called an iterative loop.
##### Syntax:
            for variable in sequence
                        statement(s)

```python
l = ["Mike", 10.2, 1980]
```

```python
for i in l: # Here i represent the iteration. We can replace with any character. So, for first iteration i represent "Mike", second iteration 10.2 and third itertaion 1980.
    print(i)
```

```python
for x in [1, "How", "Python"]: # We can directly the input also.
    print(x)

```
