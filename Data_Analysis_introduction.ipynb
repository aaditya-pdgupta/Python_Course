#### Data Analysis 
##### Data Analysis is the process of collecting, cleanin, transforming and interpreting data to extract useful insights and support decision-making.
##### The steps are :
##### 1. Collection of data.
##### 2. Processing of data.
##### 3. Comparison of data.
##### 4. Inspection of data.
##### 5. Cleaning of data.
##### 6. Summarization of data.

#### Library 
##### A library is a collection of pree-written code(functions, classes and modules) that helps programmers to perform common tasks without writing code from scratch. Some of them are :
##### 1. Scientific Libraries
##### Used for mathematica, scientific and engineering computations.
##### i.NumPy
Numerical computing, arrays, matrices, linear algebra etc.
##### ii.SciPy 
Advanced scientific calculations, signal processing, optimization, statics etc.
##### iii. Statsmodels
Statical modeling, Hypothesis testing, regression analysis etc.
##### iv.  SymPy
Symbolic mathematics, algebra, calculus, equations.

##### 2.Visualization Libraries
##### Used for graphical representation of data.
##### i.Matplotlib
Basic plotting library, Line, bar, pie, histogram plots.
##### ii.Seaborn
Statistical data visualization, Heatmaps, boxplots, distribution plots.
##### iii. Plotly
Interactive and dynamic graphs, Dashboards and web-based visuals.

##### 3. Algorithm Libraries
##### Used for machine learning, data mining, and AI.
##### i.Scikit-learn
Machine learning algorithms, Classification, regression, clustering.
##### ii.TensorFlow
Deep learning algorithms, Neural networks.
