#### Quantifiers
##### Quantifiers tell regex how many times a character or group should repeat. Example idea: Once, many times, optional and exact number.
##### The 5 main quantifiers:
##### 1. *---> 0 or more times,
##### 2. +---> 1 or more times.
##### 3. ?---> 0 or 1 time.
##### 4. {n} --> exactly n times.

```python
import re
```

```python
message = "The current Python version is 3.13. Other previous are 3.12, 3.11, 3.10."
```

```python
#pat = r"[a-z][a-z][a-z][a-z]" So instead of doing this we can do.
pat = r"[a-z]{4}" 
```

```python
match = re.search(pat, message)
```

```python
print(match)
```

```python
pat = r"[A-Z][a-z]{5}" # Here the 5 denotes the las [a-z].
```

```python
match = re.search(pat, message)
```

```python
print(match)
```

```python
pat = r"[A-Z][a-z]{2,5}" # Here 2 means there shouldl be2 minimum character  
```

```python
match = re.search(pat, message)
```

```python
print(match)
```
