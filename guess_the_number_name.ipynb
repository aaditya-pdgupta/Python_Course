##### Create a simple number guessing game.
##### The user gets 110 chances to guess a number.
#####  If the user guesses the number before 10 chances, stop asking the number from the user, say Congrats adn end the game. If the user never guesses the number , ask them 10 times and then end the game.

```python
print("Welcome to the number guessing game. We have a number that needs to be guessed. You have 10 chances.\nThe secret number is between 1 and 50.")
```

```python
correct_num = 27
number_of_times = 0
gussed = False
```

```python
while number_of_times <= 10:
    user_number = float(input("Enter your guessed number"))
    if user_number == correct_num:
        print("Congrats your guess is correct!!!")
        gussed = True
        break
    elif user_number < 27:
        number_of_times += 1
        print("Your guess is wrong! Try bit higher")
    elif user_number > 27:
        print("Your guess is wrong! Try bit lower")

if gussed:
    print("The number was 27. Game Over!!!")
else:
    ("Try again!!!")
```
