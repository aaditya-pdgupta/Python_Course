##### Create a simple number guessing game.
##### The user gets 110 chances to guess a number.
#####  If the user guesses the number before 10 chances, stop asking the number from the user, say Congrats adn end the game. If the user never guesses the number , ask them 10 times and then end the game.

```python
print("Welcome to the number guessing game. We have a number that needs to be guessed. You have 10 chances.\nThe secret number is between 1 and 50.")
```

```python
correct_num = 25
attempts = 10
num = 0
guess = False
```

```python
while num < 10:
    print(f"You have {attempts} attempts remaining.")
    user_num = float(input("Enter your guess:"))
    if user_num == correct_num:
        print("Congrats, Your guess is right, Genius!!")
        guess = True
        break
    else:
        if user_num < correct_num:
            higher_lower = "higher"
        else:
            higher_lower = "lower"
        print(f"Your guess is wrong, Try {higher_lower} number")
    num += 1
    attempts -=1

if guess == False:
    print("Bad luck! You exhausted all your attempts and couldn't guess the number.")

print(f"The secret number is {correct_num}, Game Over!!!")
            
            
                      
                    
                    
                      
    
```
