##### sub()
##### sub() is used to replace parts of a string that match a regular expressions.

##### Syntax:
##### re.sub(pattern, replacement, string)

```python
import re
```

```python
s1 = "Sunday, Monday, Tuesday, Monday, Sunday"
```

```python
pat = "Sunday"
```

```python
replacement = "Friday"
```

```python
result = re.sub(pat, replacement, s1)
```

```python
print(result) # Here we can see wherever there was Sunday it has been replaced by the Friday.
```

```python
# Suppose we want to replace only 1 than.
```

```python
pat = "Sunday"
```

```python
replacement = "Friday"
```

```python
result = re.sub(pat, replacement, s1, count = 1)
```

```python
print(result) # Here only one Sunday has been replaced and that is the first occurence.
```
