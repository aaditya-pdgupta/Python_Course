```python
import pandas as pd
import numpy as np
import seaborn as sns
np.random.seed(0)
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot(index = "month", columns = "year", values = "passengers")
```

```python
heat_map = sns.heatmap(flights)
```

```python
# Here annot Displays the numerical values inside each heatmap cell.
# Here fmt = d is used to display annotation values as integers.
heat_map_1 = sns.heatmap(flights, annot = True, fmt = 'd') 
```

```python
heat_map_2 = sns.heatmap(flights, cmap = 'YlGnBu') # cmap is used to define the color map of the heatmap.
```

```python
heat_map_3 = sns.heatmap(flights, cbar = False) #Cbar is used to display scale bar or not.
```
