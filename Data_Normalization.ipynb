#### Data Normalization
##### Data normalization is the process of scaling numeric data to a common range so that features are comparable and no single feature dominates the model. It helps to improve performance of ML algorithms (KNN, SVM, Gradient Descent, Neural Networks). The methods are:
##### 1. Mix-Max Normalization
##### 2. Z-Score Normalization (Standarization)
##### 3. Decimal Scaling
##### 4. Simple Feature Scaling
