```python
# not allowed - list, set, dict ---> mutable data types
# allowed - str, int, float, bool, tuple ----> immutable data types.
# keys of a dictionary can only be mutable data types.
# values can be any data types. 
```

```python
student1 = {"id": 1001, "name": "John", "marks": [89.5,  71.5, 81.0]} # Here there is list only as value of key marks.
```

```python
print(student1)
```

```python
print(student1["marks"]) # Here we are indexing .
```

```python
print(student1["marks"][1]) # Here we are chaining using the index to get the 71.5.
```

```python
student2 = {"id": 1002, "name": "David", "marks": {"eng": 89.5, "phy": 71.5, "chem": 81.0}} # Here the value of the marks (key) is in dict.
```

```python
# Here we want to fetch the mark of phy than.
```

```python
print(student2["marks"]) # Here we have fetched the marks.
```

```python
print(student2["marks"]['phy']) # So we have fetched here the phy marks.
```

```python
# fetching all the key. There is function keys().
```

```python
print(student2.keys()) # Here we have keys . The eng, phy are the value.
```

```python
print(student2.keys(), type(student2.keys()))
```

```python
#WE can fetch the value using the value().
```

```python
print(student2.values(), type(student2.values()))  # HEre all the value has been printed.
```

```python
#items(). If we want to fetch the pair that is key and value together.
```

```python
print(student2.items(), type(student2.items())) # We can see the value and key pair.
```
