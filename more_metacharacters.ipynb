#### ^ (Caret Symbol)
##### It has two different meanings depending on where it is used.
##### 1. ^ at the START of a pattern means start of string.
##### 2. ^ INSIDE square brackets [ ] menas NOT / negation

```python
import re
```

```python
s1 = "Python is a programming language."
```

```python
pat = r"[a-z]{8}"
```

```python
match = re.search(pat, s1)
```

```python
print(match) # Here we tried to match with the language but it get matched with the programm.
```

```python
# Here is we can use the carat ^.
```

```python
pat = r"^[a-z]{8}"
```

```python
match = re.search(pat, s1)
```

```python
print(match)  # So here we are getting none because there is no pattern in the beginning that begin acc to pat.
```

```python
# $  matches the end of the string.
```

```python
pat = r"[a-z]{8}$"
```

```python
match = re.search(pat, s1)
```

```python
print(match) # So here is no string that ends with the c
```
