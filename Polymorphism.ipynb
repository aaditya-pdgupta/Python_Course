#### Polymorphism 
##### Polymorphism means "many forms."
##### In OOPs, it allows one method or function name to perform different actions depending on the object that calls it.

```python
a = 5 # Here the + is used to add the integers.
b= 6
print(a+b)
```

```python
s1 = "Hello" # Here the same + is used for joining the string.
s2 = "BYE"
```

```python
print(s1+s2)
```
