```python
def greeting_someone(): # It is my function. We do not see output because we have defined a function but not called it.
    print("Hello, Good Morning!")
    print("It's a beautiful day.")
```

```python
# Calling the function
greeting_someone() # No we are geting the output. 
greeting_someone()
greeting_someone() # We can call multiple time. Each calling gives each output.
```

```python
def greeting(name):
    print(f"Hello {name}, Good Morining!!")
    print("It's beautiful day")
```

```python
greeting("Mark") # Calling the function just changing the arguments or parameter .
greeting("John")
greeting("Carot")
```

```python
def even_odd(number):
    if number % 2 == 0:
        print(f"{number} is Even.")
    else:
        print(f"{number} is Odd.")
```

```python
even_odd(10) # HEre we are just changing the number and an dcalled it.
```
