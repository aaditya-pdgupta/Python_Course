##### Types of Modules in Python:


#### 1. Built in Modules
#####    These are already available in Python like math, random, datetime, sys etc.

```python
#Syntax: import module_name
#Syntax for importing only few functions/variables: from module_name f1, f2, ...
```

```python
import math
```

```python
# Calculate square root of a number
```

```python
print(math.sqrt(5)) # module.function_name(arg1, arg2..)
```

```python
num = 4
```

```python
output = math.sqrt(num)
```

```python
print(f"The square root of the {num} is {output}")
```

```python
# Calculate the area of circle.
```

```python
r = float(input("Enter the radius:"))
```

```python
area = math.pi * r ** 2
```

```python
print(f"The area of circle is {area:.2f} cmÂ².")
```
