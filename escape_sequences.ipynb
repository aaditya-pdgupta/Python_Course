###  Escape sequence 
Escape sequences are special characters used inside strings to represent things thart are hard to type directly like newlines, tabs etc

Some of them are :

#### \n for new line
#### \t for Tab space
#### \\ for backslash
#### \' for single quote
#### \" for double quote
#### \r for carriage return (moves cursor to start of line)
#### \b backspace
#### \f for form feed(new page)
#### \a for Bell sound(beep)

```python
#\n
```

```python
print("Hello everyone. \nHow are you?") # Here the \n breaks the line into new line
```

```python
#\t
```

```python
print("John \t20") #Here the gap between the word is more. Here \t gives tab option.
```

```python
print("new\\old") #Here we have printed.
```
