##### Frozen set in Python
##### A frozen set is just like a normal set, but it immutable. That means:
##### 1. We cannot add elements.
##### 2. We cannot remove elements.

##### But we can do:
##### 1. Union
##### 2. Intersection
##### 3. difference
##### 4. symmetric differences

```python
s1 = {1, 2,3 ,4}
```

```python
s1.add(-1)
```

```python
print(s1) # Here we are able to add the -1 which proves the sets are mutable.
```

```python
# Frozen set---Immutable set
```

```python
fs1 = frozenset({10, 20, 30})
```

```python
print(fs1, type(fs1)) # Here the output is with the prefix called frozenset.
```

```python
fs1.add(40) # So it clarify the frozen set can not be added.
```

```python
fs2 = frozenset({3, 4, 5, 7})
```

```python
print(fs1.union(fs2)) # So the union is working here.
```

```python
print(fs1.union(fs2))
```
