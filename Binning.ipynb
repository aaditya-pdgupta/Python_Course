#### Binning (Data Discretization)
##### Binning is a data preprocessing technique in  which continuous numerical data is converted into discrete intervals (bins).
##### Example:
##### Age values → [18, 22, 25, 30, 35]
##### After binning → Young, Adult, Senior

#### Why Binning Is used:
##### 1. Reduces noise in data.
##### 2. Makes pattern easier to understand.
##### 3. Useful in machine learning and signal processing.
##### 4. Helps in data visualization.

```python
import pandas as pd
```

```python
df = pd.read_excel("Numeric.xlsx")
```

```python
df.head()
```

```python
# Numeric to Categorical
```

```python
import numpy as np
```

```python
bins = np.linspace(min(df["Length"]),max(df["Length"]),4) # It generates evenly spaced numbers between a start and end value.
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
df["Length-binned"] = pd.cut(df["Length"], bins, labels = group_names, include_lowest = True)
```
