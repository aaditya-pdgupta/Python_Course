#### Local Variable 
##### A local variable is a variable defined inside a function. It can be used only inside that function.

#### Global Variable
##### A global variable is a variable defined outside all functions. It can be acccessed inside and outside functions.

##### Note:
#####  If a variable name is same, Python gives priority to local variable.

```python
n = 1 # It is the global variable.
def fn(): # We are able both variable.
    n = 5 # It is the local variable.
    print("in",n)
fn()
print("out",n)
```

```python
n = 1
def fun():
    global n # Here although we have given the global variable n = 1, but using global inside assign the global variable the value of local variable.
    n = 5
    print("in",n)
fun()
print("out",n)
```
