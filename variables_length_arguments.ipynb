##### Variable-length Arguments 
##### Variable-length arguments are used when we don't know how many argument will be passed to a function. There are two types: 

Tricks :School Bag

*args → books thrown into the bag (no labels)

**kwargs → labeled items (name = book, color = cover)


##### 1. *args (Non-keyword Variable-length Arguments)
##### ----> Stores multiple values in a tuple.
##### ----> Used when arguments are passed without names.

```python
def add(*args):
    return sum(args)
```

```python
result = add(1,2,3) #We can add as much as number we want.
```

```python
 print(result) # If there is no argument it will give blank so we dont need to worry. It exhaust all the positional arguments.
```

```python
def add(*nums): # It is not compulsory to use the arg we can use any in the bracket arguments. But arg is standard way.
    return sum(nums)
```

```python
res = add(1,2,3)
```

```python
print(res)
```

```python
def student_details(sid,name, *marks): # Here we dont know many subject marks are there for the student so we used *.
    percent = sum(marks) / len(marks)
    print(f"{name} with id {sid} secured {percent}")
    return percent
```

```python
result = student_details(10111, "John", 55, 55, 55, 33) # We can add as much as sub we can
```

```python
print(result)
```

```python
def student(roll, name, *marks):
    if len(marks) == 0:
        print(f"{name} with enrollement no: {roll} is fail!!")
    else:
        percent = sum(marks) / len(marks)
        print(f"{name} with enrollement no: {roll} secured {percent:.2f}")
```

```python
student(101, "XYZ", 55, 20,100) # Here for Alice is fail because we have not given the any value for the marks.
student(101, "John", 55, 90,90)
student(101, "Henry", 55, 50,99)
student(101, "Alice")
```

##### 2. **kwargs (Keyword Variable-length Arguments)
##### ----> Stores arguments in a dictionary.
##### ----> Used when arguments are passed as key = value.

```python
def func(**kwargs):
    print(kwargs)
```

```python
func(x=10, y=20) # Here the output is in dictionary followed by the key and value.
```

```python
func() # If we do not give any argument. It will give the empty dictionary.
```

```python
def std_details(roll, name, **marks): # It ** should be the last arguments. We can not use the * after **.
    if len(marks) == 0:
         print(f"{name} with enrollement no: {roll} is fail!!")
    else:
        percent = sum(marks.values()) / len(marks)
        print(f"{name} with enrollement no: {roll} secured {percent:.2f} %")
```

```python
std_details(230,"John", sub1=78.5, sub2=95.5)
std_details(231,"Rama", sub1 =89, sub2 = 66, sub3= 99)
```

```python
def std_details(roll, name, *extra, **marks): # It ** should be the last arguments. We can not use the * after **.
    if len(marks) == 0:
         print(f"{name} with enrollement no: {roll} is fail!!")
    else:
        percent = sum(marks.values()) / len(marks)
        print(f"{name} with enrollement no: {roll} secured {percent:.2f} %")
    print(f"{name}, does {extra}")
```

```python
std_details(230,"John",'football', sub1=78.5, sub2=95.5) # There is comma after football because the tuple comma with comma.
```
