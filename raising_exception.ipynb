#### Raising an exception
##### Rasising an exception in Python means manually generating an error using the raise keyword when a specific comdition occurs.
##### Syntax:
   ##### raise ExceptionType("Error message")

```python
salary = float(input("Enter salary: "))
```

```python
if salary < 0: # Here we are getting the Error As ValueError.
    raise ValueError("Salary can not be negative or zero")
else:
    print(f"Your salary is {salary}")
    
```
