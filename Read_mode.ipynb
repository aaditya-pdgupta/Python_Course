#### Reading a file in Python
##### Reading a file means retrieving data stored in a file using Python.
##### Syntax is "r".


```python
file = open("Practice.txt", "r") # It reads the contents of the file as str.
```

```python
content = file.read(10) # Here we are printing the first 10 character.
```

```python
file.close()
```

#### Readlines()
##### readlines() is a method used to read entire file line by lines at once and store them in list.

```python
f2 = open("file1.txt", "r")
```

```python
 lines = f2.readlines()
```

```python
print(lines) # Here we are getting in the list and \n is included in it.
```

```python
for line in lines:  # Here we are printing line by line.
    print(line.rstrip())
```
