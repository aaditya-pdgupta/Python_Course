#### Continue.
##### 1.Skips the current iteration.
##### 2. Loop continuees with next value.

```python
for num in range(10):
    if num % 3 == 0: # If the number is divisible by 3 it will skip it.
        continue
        
    print(num)
```

```python
for i in range(1,10): 
    if i == 3: # Here it check whether the i is equal to 3. If it is, then it will skip and continue from next.
        continue
    print(i)
    
```

#### Break
##### 1. stops the loop completely.
##### 2. control comes out of the loop.

```python
for num in range(1, 10): # Here when the num = 3 the loop stop and come out of it. No further iteration continues.
    if num % 3 == 0:
        break
        
    print(num)
```

```python
for i in range(1,10): 
    if i == 5: # Here it check whether the i is equal to 3. If it is, then it will skip and continue from next.
        continue
    print(i)
```
