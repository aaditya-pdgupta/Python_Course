### Fstring 
f-string stands for formatted string literal

```python
name = "John"
```

```python
age = 20
```

```python
language = "Python"
```

```python
hours = 3
```

```python
# John is 20 years old. 
```

```python
print(name, "is", age, "years old.")
```

```python
# John is 20 years old. He studies Python 3 hours a day.
```

```python
print(name, "is", age, "years old. He studies", language,hours, "a day.")
```

```python
# Using f-string. Put the curly bracket whereever there is variable.
```

```python
print(f"{name} is {age} years old. He studies {language} {hours} hours a day.") # Here f is used before.
```

```python
sub1 = 78
```

```python
sub2 = 87
```

```python
sub3 = 83
```

```python
#John scored 
```

```python
print(f"{name} scored {sub1 + sub2 + sub3} marks in total.") # Here we did not create any variable that sum all three subject we just did it in print directly.
```

```python
percent = (sub1 + sub2 + sub3)/3
```

```python
print (f"{name} got {percent}%")
```

```python
print (f"{name} got {round(percent,2)}%") # Here we have just round off the percent.
```
