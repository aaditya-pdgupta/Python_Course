#### Operator Overloading
##### Operator Overloading is a feature of Object-Oriented Programming that allows operators to have different meanings when applied to user defined objects. It means using the same operator for different purposes.

```python
a = 20
b = 20
print(a + b) # a.__add__(b)
```

```python
class A:
    def f1(self, val):
        pass
```

```python
obj = A()
```

```python
obj.f1(20) #Here A.f1(obj, 20)
```

```python
s1 = "Hello"
s2 = "Hi"
print(s1 + s2) # str.__add__(s1, s2)
```

```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
```

```python
    def area(self):
        return self.length * self.breadth
```

```python
r1 = Rectangle(length = 5, breadth = 4)
r2 = Rectangle(length = 8, breadth = 6)
```

```python
print(r1.area())
```
