#### Abstract Class
##### An abstract class is a class that cannot be instantiated (object can not be created) and is used to define a blueprint for other classes. It contains abstract methods (methods without implementation).

```python
class Square:
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2
```

```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length 
        self.breadth = breadth 

    def area(self):
        return self.length * self.breadth
    
```

```python
sq1 = Square(10)
```

```python
print(sq1.area())
```

```python
r1 = Rectangle(length = 5, breadth = 3) # Here all the set need area. So we can make it by abstract method.
```

```python
print(r1.area())
```
