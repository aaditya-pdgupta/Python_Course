#### String_cases_in_Python
##### In python, string "case" refers to whether letters are uppercase or lowercase. Python provides several methods to check or change case.

##### 1. upper()--- convert all letters to uppercase.
        syntax: .uppercase()
##### 2. lower()---- convert all letters to lowercase.
        syntax: .lowercase()
##### 3. title()---- Convert the first letter of each word to uppercase, rest lowercase.
        Syntax: .title()
##### 4. capitalize()---- Only the first character of the string becomes uppercase.
        sntax: .capitalize()
##### 5.swapcase()---- convert uppercase to lowercase and vice versa.
        syntax: .swapcase()
##### 6. isupper()------True if all letters are uppercase.
#####    islower()------True if all letters are lowercase.
#####    istitle()------True if string is in title case.

```python
s1 = "Python"
```

```python
print(s1.upper()) # Here the entire word is converted in the upper case.
```

```python
s2 = "Python3.13"
```

```python
print(s2.upper()) #The uppercase works for the string not for integer.
```

```python
print(s1.lower()) # Here all P is upper case only so now using this all the word is converted in the lowercase.
```

```python
s3 =  "We are learning Python"
```

```python
print(s3.upper()) # All the sentence is converted into the uppercase.
```

```python
print(s3.lower()) #All the sentence is converted into the lowercase.
```

```python
print(s3.title()) # Al
```
