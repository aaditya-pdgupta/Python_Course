#### Method Overriding 
##### Method overriding occurs when a child class provides its own implementation of a method that is already defined in the parent class.

##### The method name and parameters remain the same but the behavior changes.

```python
class Employee:
    def working_hours(self):
        return 45
```

```python
class Intern(Employee):
    def working_hours(self):
        return 40
```

```python
i1 = Intern()
```

```python
print(i1.working_hours()) # Here we are getting the working hours of intern.
```
