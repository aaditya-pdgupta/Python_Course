#### Dictionary in Python
##### A dictionary is a collection of key-value pairs. Each value is accessed using its key, not an index. It is represented as dict.
##### Syntax:
                my_dict = {
                     "name": "Aadi",
                      "age": 20,
                       "country": "India"
                    }


```python
#{key1 : value1, K=key2 : value2, ....}
```

```python
groceries = {"milk": 60, "biscuits": 20, "rice" : 90, "bread" : 30}
```

```python
print(groceries)
```

```python
type(groceries) # Here the data type is dictionary.
```

```python
len(groceries)  # We can use the len function to find the key and value pair.
```

```python
groceries[3] # The index is not possible in dictionary to fetch the element.
```

```python
print(groceries["milk"]) # We can use the key instead of index to fetch the value.
```

```python
groceries["milk"] = 65 # We can do indexing to updated the value. We use the key.
```

```python
print(groceries) # Here the value has been changed. Hence the dictionary is mutable.
```

```python
print(groceries["eggs"]) # There is no egg present in dictionary as key so it is showing error.
```

```python
groceries["eggs"] = 6 # The dictionary has been updated with egg. So it is a add function to add key-value pair. If the key is already present than it update the existing the key 
```

```python
print(groceries)
```
