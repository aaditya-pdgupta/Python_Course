#### Returning Values in Python (return)
##### In Python, a function returns a value using the return keyword. The returned value can be stored, printed or used elsewhere in the program.
##### Syntax:
            def function_name():
                return value

def even_odd(number):
    if number % 2 == 0:
        print(f"{number} is Even.")
    else:
        print(f"{number} is Odd.")

```python
even_odd(9) # Here it is not returning anything. It is just giving the anser based on the arguments.
```

```python
result = even_odd(9) 
```

```python
print(result) # The function does not send any value back. It only displays the result like above. We can not use the result for further calculation.
```

##### Now let use the return and see the result.

```python
def even_odd(number):
    if number % 2 == 0:
        return"Even"
    else:
        return"Odd"
```

```python
result = even_odd(9)
```

```python
print(result) # Now we are geting the result.
```
