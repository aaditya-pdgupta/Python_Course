```python
# Write a program to simulate a roll of dice.
# A dice has 6 faces with numbers 1 to 6 written on them.
# The program should randomly print a number between 1 and 6.
```

```python
import random
```

```python
print("Welcome to the game of rolling a dice.")
```

```python
while True:
    choice = input("Press 'Enter' to roll the dice or 'q' to quit.")
    if choice =='q':
        print("Thanks for playing the game,  have a good day!!")
        break
    elif choice == "":
        number = random.randint(1,6)
        print(f"Your number is {number}")
    else:
        print("Invalid input!!!")
```

```python
num = 0
attempt = 3
rolled = False
```

```python
while num < 3:
    print(f"The number of the attempt remaining is {attempt}")
    user_num = input("Press 'Enter' to continue game or 'q' to quit")
    if user_num == 'q':
        print("Thanks for visiting")
        rolled = True
        break
    elif user_num == '':
        number =  random.randint(1,6)
        print(f"Your lucky number is {number}")
    else:
        print("Invalid input!!!")
    num += 1
    attempt -= 1
if rolled == False:
    print(" The attempt is over, visit again!!!")

    
```
