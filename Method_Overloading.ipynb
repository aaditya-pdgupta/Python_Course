#### Method Overloading
##### Method overloading means having multiple methods with the same name but with different parameters.

```python
class A:
    def add(self, num1, num2): # Here in this add function we have argument num1 and num2.
        return num1 + num2 
```

```python
    def add(num1, num2, num3):
        return num1 + num2 + num3  # Here we have same function with the three argument.
        
```

```python
obj = A()
```

```python
print(obj.add(num1 = 10, num2 = 2))
```
