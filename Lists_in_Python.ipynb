#### List in Python
##### A list is a data structure in Python used to store multiple items in a single variable. Items can be of any data type (integer, strings, floats, other lists, etc.)

##### Syntax:
        my_list = [item1, item2, item3,.....]

```python
days_of_week = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"] # Here also indexing is done as the string we did. So here Mon is o and Tue is 1 and later on.
```

```python
print(days_of_week[0]) # Here 0 is for mondat so output is Mon.
```

```python
print(days_of_week[4])# Here 4 is for Friday so output is Fri.
```

```python
print(days_of_week[-1]) # Here using the negative indexing we printed the output.
```

```python
print(f"Last days of the week is {days_of_week[-1]}") # Here using f string we printed the output.
```

```python
#  We can print the length of list same as the length of the string.
```

```python
print(len(days_of_week)) # Here there are althogether 7 days so the length is 7.
```

```python
print(f"There are altogether {len(days_of_week)} days in a week") # We can print this output using the f string.
```
