#### Membership operator
##### in
##### Checks if a value is present inside a sequence

```python
s1 = "Python is fun"
```

```python
print("Python" in s1) # Here the python is present so it give  the boolean as True.
```

```python
print("i" in s1) # Here we can check if the single character is present or not.
```

```python
print("z" in s1) # Same here aslo there is no z in s1 so the value is false.
```

```python
print("Java" in s1) # There is no Java in s1.
```

##### There is also another membership operator called as not in. It is just opposite of the in operator.

```python
print("Java" not in s1) # there is no Java in s1 so the value is true.
```

```python
print("P" not in s1) # There is P in the s1 so the output is false.
```

#### replace()
##### replace() is used to replace a part of a string with another string
#### syntax
      string.replace(old, new, count)
#### old = the substring we want to replace 
#### new = the substring we want to insert
#### count = number of times to replace (optional)

```python
text = "We are learning Python."
```

```python
print(text.replace("Python", "Java")) # The input will not change just the output is new.
```

```python
print(text) # See here the input is same.
```

```python
text = text.replace("Python", "Java") # here if we want to save the replaced than we can do this.
```

```python
print(text) # Now the output ischanged and stored in the text.
```

```python
print(text.replace("e", "E")) # Here all the small e changed in the Capital E.
```

```python
print(text.replace("e", "E", 2)) # Here the first two e are only changed.
```

```python
s7 = "Hello World"
```

```python
print(s7.replace("World", "")) # To replace any specific character or word just replace it with an empty string.
```

#### Counting substring from a string
Counting means finding how many times a substring, word or character appears inside a string.
##### syntax
           string.count(substring)

```python
n1 = "We are learning Python. Python is fun"
```

```python
n2 = "Python"
```

```python
n3 = n1.count(n2)
```

```python
print(n3) # Here the number of the Python present in the n1 is 2.
```

```python
print(f"Occurence of the {n2} is {n3}")
```

```python
n4 = n1.count("e") # Here the presence of e in the 3 time.
```

```python
print(n4)
```

```python
n5 = n1.count("learn")
```

```python
print(n5) # Here the occurence of the learn is only one time.
```
