#### Suppoose I need to print this pattern.
##### *
##### **
##### ***
##### ****
##### *****

```python
for i in range(1, 6): # Here it will control how many rows will print.
    for j in range(1, i+1): # Here this decide how many times the star should be printed.
        print("*", end = " ") # Here end adds a space after
    print()  # moves to the next row
    
```

```python
for i in range(1, 10):
    for j in range (2, 2*i+1):
        print("AADI", end = " ")
    print()
```
