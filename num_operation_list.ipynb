```python
numbers = [10, 4, 5.5, 6, 3]
```

```python
# Smallest number in the list.
```

#### min()
##### It helps to find the minimum value in a list in Python. 
#### Syntax: 
        min(list_name)

```python
min(numbers)
```

```python
print(min(numbers))  # Here it gives the min number as 3.
```

#### max()
##### max() helps to find the maximum value in a list in Python.
##### syntax:
            max(list_name)

```python
print(max(numbers)) # We have the highest number here is 10. So the output is 10.
```

```python
print(f"Biggest number is {max(numbers)}")
```
