#### w mode
##### "w" stands for write mode.
#### Function:
##### 1. Creates a new file if it does not exist.
##### 2. Overwrites the file if it already exists.
##### 3. Writes data in text format.

```python
file = open("file1.txt", "w")
```

```python
file.write("This file is overwritten using "w" mode in Python.\n")
```
