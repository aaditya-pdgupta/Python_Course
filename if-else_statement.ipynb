#### if-else statement in Python
##### if-else is used when you want to run one more block if the condition is True, and another block of the condition is false.
##### Syntax:
            if condition:
                statement1
            else:
                statement2
    

```python
age = float(input("What is your age?")) # We are giving the input as age of a person
```

```python
if age >= 18: # Here a conditio is that a person must be equal or above the 18 to cast vote. .
    print("Congrats!!! You are eligible for casting the vote.")
else:
    print("A few more years before you can cast vote.")
    
```

```python
# Write a program to print if a number (int) is odd or even.
#even --- when the number is divisible by 2 and remainder is 0.
#odd -- when the number is not divisibleby 2 and remaineder is not 0.
```

```python
num = int(input("Enter an integer:"))
```

```python
if num % 2 ==0: # The % gives the remainer. If the remainder is 0 than the number is even and if not than it is odd.
    print("The given integer is even")
else:
    print("The given integer is odd")
```
