#### Class Variable
##### A class variable is a variable thar is :
##### 1. Shared by all objects (instances) of a class.
##### 2. Defined inside inside the class but outside any method.

```python
class Student:
    college_name = "ABC College" #  Here we have two class variable.
    departments = ["Arts", "Commerce", "Science"]

    def __init__(self, name, roll): # We have three instance method.
        self.name = name
        self.roll = roll

    def study(self, n_hours):
        print(f"The student studies for {n_hours} hours a day!")
        
    def sports(self, sports_name):
        print(f"The student plays {sports_name}")
```

```python
student1 = Student(name= "John", roll = 1021)
```

```python
print(student1.name)
```

```python
help(Student)
```

```python
print(student1.__dict__) # Here dict show the instace variable not the class variable.
```

```python
print(student1.departments) # We can print the class variable using the object name.
```

```python
print(student1.college_name)
```

```python
print(Student.departments) # We can print the class variable using the Class name also..
print(Student.college_name)
```
