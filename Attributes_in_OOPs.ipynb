#### Attributes in OOPs
##### Attributes are variables that belongs to a class or an object. They store data/ properties of an object. Attributes are accessed using the dot (.) operator.

```python
class Student:
    pass
    
```

```python
student1 = Student()
student2 = Student()
```

```python
student1.name = "John" # Here they are the attributes. It is the attributes for the object student1.
student2.roll = 1001
```

```python
print(student1.name)
```

```python
print(student2.roll)
```

```python
print(student2.name) # Here if we checking for the student2 we are getting the attribute error.
```

```python
help(Student)
```

```python
print(student1.__dict__) # It will store the variable which we have created.
```

```python
print(student2.__dict__) 
```
