#### Boxplot
##### A box plot is a graphical representation use to show the distribution of numerical data based on five summaryy statisics.

#### What does a Box Plot Show?
##### 1. Minimum
##### 2. Q1(25%)
##### 3. Median (Q2/50%)
##### 4. Q3(75%)
##### 5. Maximum 
##### 6. Outliers(if any)

```python
import seaborn as sns
sns.boxplot(x = "Length", y = "E", data = df)
```

```python
#SCATTERPLOT
import matplotlib.pyplot as plt

x = df["Length"] 
y = df["E"]
plt.scatter(x,y)
plt.title("Scatterplot along the x and y axis")
plt.xlabel("Length")
plt.ylabel("Width")
```

##### groupby()
##### groupby() is used to group data based on one or more columns and then perfor calculations (like sum, mean, count etc) on each group.

```python
import numpy as np
```

```python
grp = df.groupby("A")
grp.aggregate(np.sum)
```

```python
grp = df.groupby(["A", "B"])
grp.aggregate(np.sum)
```

```python
grp = df.groupby(["A", "B"], as_index = False) # Here it has been proper manner.
grp.aggregate(np.sum)
```

```python
grp.size()
```

```python
grp.count()
```

#### Pivot 
##### In pandas, pivot() is used to reshape data--- it converts rows in into columns.
##### It is mainly used to create a pivot cable from a DataFrame.
#### Syntax:
df.pivot(index, columns, values)


##### Parameters:
##### index ---> Rows (unique values becomes row labels)
##### columns --> Column headers 
##### values ---> Values to fill inside  the table.


```python
df_20.pivot(index = "A", columns = "B", values = "C")
```

```python
df_20.pivot(index = "S.No", columns = "A")["B"]
```

```python
df_20.pivot(index = "S.No", columns = "A", values = ["B", "C"])
```

#### Heat Map in Data  Analysis
##### A Heat Map in a data visualization technique used in data analysis to repressent numerical values using colors.
##### --> Different colors indicate different magnitudes of data.
##### --> It helps in identifying patterns, trends and correlations quickly. 

##### import pandas as pd 
##### import seaborn as sns
##### import matplotlib.pyplot as plt

```python
df = pd.DataFrame({ 'Math': [80, 85, 90, 70], 'Science': [78, 88, 92, 68], 'English': [75, 82, 89, 65]}) 
```

```python
corr = df.corr() # It is used to calculate the correlation between numerical columns in a DataFrame.
```

```python
sns.heatmap(corr, annot = True)
plt.show()
```
