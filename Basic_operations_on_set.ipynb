```python
nums1 = {1, 3, 2, 0, -1}
nums2 = {6, 7, 8, 9, 5}
```

#### Membership Operator
##### 1. in
##### 2. not in


```python
print(1 in nums1) # here o is present so the output is true.
```

```python
print(33 not in nums1) # 33 is not present so the output is true.
```

```python
print(nums1 + nums2) # Hence we can not perform the concatenation is not possible in set.
```

```python
print(nums1 * 3) # We can not perform repeating set.
```

```python
weekdays = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")
```

```python
set(weekdays) # We can convert the tuple into set.
```

```python
print(weekdays) # We just reassign so we need to store in some variable.
```

```python
weekdays = set(weekdays)
```

```python
print(weekdays) # Here it has been stored. But the output is unordered. Each and every time the order get changed.
```

##### Sets are mutable.

```python
set1 = {1, 2, 3}
```

```python
set1[1] = 3 # We can not do indexing because there is no position.
```

##### add(), it add element to the set. It is unordered so we can not set where the element will be added.

```python
set1.add(5)
```

```python
print(set1)  # Here it has been added in the set.
```

```python
set1
```

```python
set1.remove(1) # Here we are specifying the elemennt 1 should be removed.
```

```python
print(set1) # Here the remove is possible.
```

```python
set1.add(2) # Adding the existing element does not give error but it also does not add because set does not allow duplication.
```
