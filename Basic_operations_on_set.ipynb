```python
nums1 = {1, 3, 2, 0, -1}
nums2 = {6, 7, 8, 9, 5}
```

#### Membership Operator
##### 1. in
##### 2. not in


```python
print(1 in nums1) # here o is present so the output is true.
```

```python
print(33 not in nums1) # 33 is not present so the output is true.
```

```python
print(nums1 + nums2) # Hence we can not perform the concatenation is not possible in set.
```

```python
print(nums1 * 3) # We can not perform repeating set.
```

```python
weekdays = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")
```

```python
set(weekdays) # We can convert the set into tuple.
```

```python
print(weekdays) # We just reassign so we need to store in 
```
