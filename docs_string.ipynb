#### Doc String
##### A docstring(documentation string) is a special string used to describe what a Python module, function, class or method does.
##### Syntax
                def function_name():
                     """This is a docstring."""
                     statement

```python
def func():
    """
    This is a docstring
    We can write what the function does
    :return:None
    """
```

```python
func() # Here calling the func does not give any error mean it is correct way of writing the doc string. IT is simply commenting.
```

```python
print(help(func)) # Help function displays the doceumentation (docstrin) of func.
```

```python
def divide(num1, num2):
    """
    num1: A number to be divided (numerator)
    num2: A number that divides num1 (denominator)
    return :float
    """
    result = num1/num2
    return result
```

```python
print(help(divide))
```

```python
result = divide(10,2)
```

```python
print(result) # Here we are getting the rsult.
```

```python
result = divide(10, 0) # Here we are getting the error.
```
