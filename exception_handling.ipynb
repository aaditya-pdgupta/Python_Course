```python
try:
    with open("my_file.txt", "r") as fh:
        data = fh.read()

    print(data)
except FileNotFoundError:
    print("File that you are trying to open does not exist!")
```

```python
# else block is used after try and except. It executes only when NO exception occurs in the try block.
```

```python
try:
    with open("Practice.txt", "r") as fh:
        data = fh.read()

    print(data)
except FileNotFoundError:
    print("File that you are trying to open does not exist!")
else:
    print(data)
    
```
