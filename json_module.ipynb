#### JSON
##### JSON(JavaScript Object Notation) is a lightweight data-exchange format used to store and transfer data.

##### Why Use json Module?
##### The json module in Python is used to:
##### i> Convert Python objects ----> JSON
##### ii> Convert JSON -----> Python Objects
##### iii> Read/write JSON files

```python
import json
```

```python
students = {"student1": {"roll": 101, "name": "John", "percent": 98.5, "sports": True},
             "student2": {"roll": 102, "name": "Carol", "percent": 92.5, "sports": False},
             "student3": {"roll": 103, "name": "Alice", "percent": 80.0, "sports":True}}
```

```python
print(students)
```

```python
print(type(students))
```

```python
#dump()
```

```python
with open("student_data.json", "w") as f:
    json.dump(students,f, indent = 4)
```
