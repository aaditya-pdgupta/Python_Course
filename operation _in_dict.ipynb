```python
emp1 = {"id": 111, "name": "John", "salary": 25000}
```

```python
print(emp1.get("phone")) # Here the key phone is not present so the value is none.
```

```python
print(emp1.get("phone", 811111111)) # Here although key is not present , it gives the default value.
```

```python
print(emp1.get("id", 811111111)) # There is key id and we have given the default value too. But it will give the key value rather than the default value asid is present over there.
```

```python
# Membership operator
```

```python
print(111 in emp1) # Here the output is false although there is 111 in emp1 because 111 is value not a key. It checks the key not value.
```

```python
print("id" in emp1) # Here the output is true so it shows that it check the key not value.
```

```python
emp1["phone"] = 81111111 # We can add the key value pair like this in ictionary.
```

```python
print(emp1)
```

#### Update function

```python
sem1_marks = {"maths": 78, "eng": 88.2, "phy": 95.2}
```

```python
sem2_marks = {"chem": 85.0, "bio": 99.2}
```

```python
sem1_marks.update(sem2_marks) # Here we are updating the sem1 marks with sem 2 marks.
```

```python
print(sem1_marks) # The sem1 marks has been updated.
```

```python
groceries_1 = {"milk": 60, "rice": 100, "biscuits": 20}
```

```python
groceries_2 = {"rice": 110, "bread": 88}
```

```python
groceries_1.update(groceries_2) # Here the new price of the rice is updated with new price while there was no bread in groceries_1 so it has been updated with the bread from groceries_2.
```

```python
print(groceries_1)
```
