```python
countries = ["Nepal", "United States", "Australia", "Ireland", "Srilanka", "Iceland", "Cuba", "Iran", "Poland"]
```

```python
counter = 0
```

```python
# Count all the countries which are starting with N.
```

```python
for country in countries:
    if country[0] == "N":
        counter = counter + 1
        
print(f" The numberof countries starting with N is {counter}")
```

```python
# There is also another way to do with start.with
```

```python
count = 0
```

```python
for country in countries:
    if country.startswith("N"):
        count = count + 1
        
print(f" The number of countries starting with N is {counter}")
```
