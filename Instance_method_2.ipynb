```python
class Student:
    """
    This is a class Student to manage student info and activities
    """
    def study(self):
        print(f"self is: {self}")
        print("The student studies for 3 hours per day!!")
```

```python
student1 = Student()
```

```python
print(f"The object: {student1}")
```

```python
student1.study()
"""
When we call an instance method using the object/instance of the class, Python passes the object itself as the first argument that method. That first argument is by standard is self
"""
```

```python
student2 = Student() # Here each time different object will be passed.

print(f"The object: {student1}")
student1.study()
```

```python
class Student:
    """
    This is a class Student to manage student info and activities
    """
    def study(self, n_hours): # Here we are adding the n_hours for the different students.
        print(f"self is: {self}")
        print(f"The student studies for {n_hours} hours per day!!")
        
    def sports(self, sports_name):
        print(f"Students plays {sports_name}")
        
```

```python
student1 = Student() 
print(f"The object: {student1}")
student1.study(2)
student1.sports("Cricket")
```

```python
student2 = Student()

print(f"The object: {student2}")
student2.study(3)
student2.sports("Football")
```
