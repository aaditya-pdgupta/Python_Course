#### The with statement
##### The with statement is used to open a file and automatically close it after use ---even if an error occurs. No need to write close().

##### Syntax:
#####        with open("filename.txt", "mode") as file:
                   # file operation

```python
file = open("Practice.txt", "r")
```

```python
contents = file.read()
```

```python
file.close()
```

```python
print(contents)
```

```python
# This are the manual way of closing the file.
```

```python
with open("Practice.txt", "r") as f:
    content = f.read()
```

```python
print(content) # Here we have not closed any file.
```

##### Advantages:
##### 1. Automatically closes the file.
##### 2. Makes code shorter and safer.
##### 3. Prevents file memory leaks.
##### 4. Works with all file mods(r, w, a, etc)

```python
with open("With_Statement.txt", "x") as fh:
    fh.write("This is the file created using the with statement.\n")
    fh.write("Bye!")
    
```
