#### The with statement
##### The with statement is used to open a file and automatically close it after use ---even if an error occurs. No need to write close().

##### Syntax:
#####        with open("filename.txt", "mode") as file:
                   # file operation

```python
file = open("Prac.txt", "r")
```

```python
contents = file.read()
```

```python
file.close()
```

```python
print(contents)
```
