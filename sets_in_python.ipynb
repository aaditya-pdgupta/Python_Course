#### Set in Python
##### Characteristics
##### 1. Unordered
        -Elements have no fixed position.
        -Output order may chnage each time.
##### 2. No Duplicates
        -Automatically removes repeated values.
##### 3. Mutable ( Changeable)
        -You can add or remove elemnets.
##### 4. But elements must be Immutable
        -Numbers, strings, tuples allowed.
        -Lists/dictionaries not allowed inside a set.
##### We use curly bracket for this {}.

```python
set1 = {10, "Python", 2.5} 
```
