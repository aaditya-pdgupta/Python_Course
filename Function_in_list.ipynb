#### remove()
##### remove() is used to delete the first occurence of a specific value from a list.
##### Syntax:
            list_name.remove(value or element)

```python
fruits = ["Apple", "Mango", "Orange"]
```

```python
fruits.remove("Mango") # Here we are removing the specific element that is the Mango.
```

```python
print(fruits) # Mango has been deleted.
```

```python
fruits.remove("Banana") # The banana is not in the list of the fruits so it is giving the error.
```

```python
sports = ["Cricket", "Hockey", "Cricket", "Chess"] # Here we have two cricket.
```

```python
sports.remove("Cricket")
```

```python
print(sports) # Here the first cricket is removed. It mean the first occurence value will be deleted.
```

#### pop()
##### pop() is used to remove an element from a list by index.
##### If we do not give an index, it removes the last element by default.
##### Syntax :
            list_name.pop(index) # remove specific index
            list_name.pop()      # remove last item

```python
print(sports)
```

```python
sports.pop(1) # Here the index 1 represent the Cricket so it has been deleted.
```

```python
print(sports)
```

```python
num = [1, 2, 3, 4, 5]
```

```python
num.pop()
```

```python
print(num) # Above we have not used any index so it automatically deleted the last one.
```

```python
num.pop(-1) # we can use the negative indexing for pop.
```

```python
print(num)
```
