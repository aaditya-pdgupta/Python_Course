#### remove()
##### remove() is used to delete the first occurence of a specific value from a list.
##### Syntax:
            list_name.remove(value or element)

```python
fruits = ["Apple", "Mango", "Orange"]
```

```python
fruits.remove("Mango") # Here we are removing the specific element that is the Mango.
```

```python
print(fruits) # Mango has been deleted.
```

```python
fruits.remove("Banana") # The banana is not in the list of the fruits so it is giving the error.
```

```python
sports = ["Cricket", "Hockey", "Cricket", "Chess"] # Here we have two cricket.
```

```python
sports.remove("Cricket")
```

```python
print(sports) # Here the first cricket is removed. It mean the first occurence value will be deleted.
```

#### pop()
##### pop() is used to remove an element from a list by index.
##### If we do not give an index, it removes the last element by default.
##### Syntax :
            list_name.pop(index) # remove specific index
            list_name.pop()      # remove last item

```python
print(sports)
```

```python
sports.pop(1) # Here the index 1 represent the Cricket so it has been deleted.
```

```python
print(sports)
```

```python
num = [1, 2, 3, 4, 5]
```

```python
num.pop()
```

```python
print(num) # Above we have not used any index so it automatically deleted the last one.
```

```python
num.pop(-1) # we can use the negative indexing for pop hence the last element has been deleted.
```

```python
print(num)
```

#### reverse()
##### reverse() is used to reverse the order of elemnts in a list in-place (means it changes the original list).

##### Syntax:
            list_name.reverse()

```python
days_of_weeks = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

```python
print(days_of_weeks)
```

```python
days_of_weeks.reverse() # It reverse the entire list.
```

```python
print(days_of_weeks) # The output is reversed.
```

```python
nums = [1, 2, 3, 4, 5]
```

```python
print(nums)
```

```python
nums.reverse()
```

```python
print(nums)
```

#### Sorting in Python, sort()
##### sort() is used to arrange the elements of a list in ascending order. It modifies the original list(in-place).
##### It sort in ascending order by default.  
##### syntax
            list_name.sort()
##### For the desecnding order.
#####    Use reverse=True
        syntax
            list_name.sort(reverse=True)

```python
nums.sort()
```

```python
print(nums) # The list has been sorted in ascending order.
```

```python
nums.sort(reverse=True)
```

```python
print(nums) # The list has been sorted in the desecnding order.
```

```python
days_of_weeks.sort()
```

```python
print(days_of_weeks) # The days of week has been printed in ascending order.
```

```python
days_of_weeks.sort(reverse=True)
```

```python
print(days_of_weeks)  # The days of week has been printed in the descending order.
```

#### Count
##### count() is a built-in list method used to count how many times a specific value appears in a list.
##### syntax
            list.count(element)

```python
numbers = [1, 0, 3, 0, 5, 1, 0]
```

```python
print(numbers.count(0)) # Here the number of occurence of 0 is 3 times.
```

###### One basic program for integer using count().

```python
num = [1, 2, 4, 3, 2, 4, 2, 2, 3] # It is a list.
```

```python
print(f"The list of number is {num}") # It will print the list.
```

```python
item_to_count = int(input("Enter the number to be counted from the above list:")) # The user will give the number.
```

```python
c = num.count(item_to_count) # The number given by the user will be counted from the list and stored in c.
```

```python
print(f"The occurence of {item_to_count} is {c}") # The output will be printed.
```
