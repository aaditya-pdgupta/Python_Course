#### Range in for loop range()
##### The range() functions generates a sequence of numbers or integers.
##### It is commonly used with for loops to repeat actions a fixed numbers of times.
##### syntax :
               for i in range(start, stop, step):

##### Parameters:
1. starts-----> beginning value (default = 0)
2. stop-------> ending value (NOT included)
3. step-------> jump between numbers (default = 1)

               

```python
for i in range (1,10,1): #The number start from the 1 but here 10 is excluded and step is 1 mean it will jump 1.
    print(i)
```

```python
for i in range (1,11,2): # Here the progress is 2 ahead jump.
    print(i)
```

```python
# Generate even numbers between 1 and 10 (10 exclude)
```

```python
for i in range (2, 10, 2): # here starting with 2 and increment is 2 means it will continue to 8 and 10 is stop so it is excluded.
    print(i)
    
```

```python
# Generate reverse orde ----> 20 to 10 (excluding 10)
```

```python
for i in range (20,10,1): # Here no output is generated because logically doing +1 step we will never reach 10 from 20.
    print(i)
```

```python
for i in range (20, 10, -1): # Here we have made step -1
    print(i)
```

```python
# Generate reverse orde ----> 20 to 10 (excluding 10) even only
```

```python
for i in range (20, 10, -2): # Here we have made step -2 and the output is reverse even number.
    print(i)
```

```python
# From here we conclude that for increment we use + step and for decrement we use -1 step.
```
