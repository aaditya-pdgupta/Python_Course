#### File handling
##### File handling in Python is used to store, read and modify data permanently using the files.
#### Files Modes
##### r -----> read
##### w -----> write
##### a -----> append
##### rt ----> read & write

#### Text Files
##### It stores data in human readable form.  examples. .txt, .csv

#### Binary Files
##### It stores data in binary (0s and 1s). examples: images, audio, medical
