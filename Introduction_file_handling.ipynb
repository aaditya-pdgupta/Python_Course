#### File Handling
##### File handling in python is used to store, read and modify data permanently using the files.
#### Files Modes
##### r -----> read
##### w -----> write
##### a -----> append
##### rt ----> read & write

#### 1. Text Files
##### It stores data in human readable form.  examples. .txt, .csv

#### 2. Binary Files
##### It stores data in binary (0s and 1s). examples: images, audio, medical signal files etc.
