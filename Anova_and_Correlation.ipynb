#### Anova
##### ANOVA (Analysis of Variance) is a statistical test used to  determine whether there is a significant difference between the means of three or more groups. It checks whether the variation between groups is greater than the variation within groups. There are two main parameters. They are:
##### i> F-Test Score (F-value)
##### The F-Test score is the ratio of between group variance to within group variance in ANOVA. A large F-Valueu indicates a large difference between group means. A small F-value indicated little or no difference between group means.

##### ii> p-value
##### The p-value shows the probability of obtaining the observed result if the null hypothesis is true.
##### if p<0.05 ---> Reject the null hypothesis---> Significant difference exists.
##### if p>0.05 ---> Accept the null hypothesis ---> No significant difference.
