```python
import re
```

```python
s1 = "Python is a programming language."
```

```python
# [A-Z], [a-z]
```

```python
pat = r"[A-Z] [a-z]" # The r stands for raw. It tells Python: Don't treat backslashes(\) as escape characters.
```

```python
p = r"old\new"
```

```python
print(p) # Here the output is printed as backslash without breaking the line.
```

```python
match_object = re.search(pat, s1)
```

```python
print(match_object)
```
