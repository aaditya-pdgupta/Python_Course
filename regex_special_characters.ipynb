```python
import re
```

```python
s1 = "Python is a programming language."
```

```python
# [A-Z], [a-z]
```

```python
#p = r"old\new"
```

```python
#print(p) # Here the output is printed as backslash without breaking the line.
```

```python
pat = r"[a-z][a-z]" # The r stands for raw. It tells Python: Don't treat backslashes(\) as escape characters.
```

```python
match_object = re.search(pat, s1)
```

```python
print(match_object)
```

```python
# Now let use another case
```

```python
pa = r"[A-Z][a-z][a-z]" 
```

```python
match = re.search(pa, s1)
```

```python
print(match)
```

```python
# \d mean Digit and \D mean Non-Digit
```

```python
# \d matches 1  digit character. It is similar [0-9]
```

```python
s2 = "Python is a programming Language. python3.13 is the current version."
```

```python
p = r"[a-z][a-z][a-z]\d" # This mean three lowercase letters followed by one digit.
```

```python
match_obj = re.search(p, s2)
```

```python
print(match_obj) # So here the hon is followed by 3.13 .
```

```python
# \D matches 1 non-digit character. It is similar to [0-9]
```

```python
paa = r"[a-z][a-z][a-z]\D" # It is just opposite of the small d it matches with anything except numbers.
```

```python
m = re.search(paa, s2)
```

```python
print(m)
```
