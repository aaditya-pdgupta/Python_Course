#### Nested list
##### A nested list is a list that contains other lists as its elements.  It's like a list inside a list.
##### Syntax:
            

```python
l1 = [5, 1.5, "Python", True, None, [1, 2, 3], 10] # There is list inside the list.
```

```python
print(len(l1))  # We can see the length of list.
```

```python
print(l1[-2]) # If we do negative indexing we see the value.
```

```python
print(l1[-2][0]) # To fetch the element of the list inside list suppose 1 i need to fetch. I can first do the indexing for list 
```
