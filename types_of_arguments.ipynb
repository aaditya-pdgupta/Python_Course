```python
def add(a,b):
    return a + b
    
```

#### Positional Arguments
##### Passed in the same order as parameters.

```python
result = add(2,5) # The first goes to the a and the second goes to b automatically.
```

```python
print(result)
```

#### Default Arguments
##### 1.Parameter has a default value.
##### 2.Used if no value is passsed.

```python
def add(a,b = 10): # Here we are assigning the default value for the b.
    return a + b
    
```
