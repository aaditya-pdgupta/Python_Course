```python
def add(a,b):
    return a + b
    
```

#### Positional Arguments
##### Passed in the same order as parameters.

```python
result = add(2,5) # The first goes to the a and the second goes to b automatically.
```

```python
print(result)
```

#### Default Arguments
##### 1.Parameter has a default value.
##### 2.Used if no value is passsed.

```python
def adds(c,d = 10): # Here we are assigning the default value for the d.
    return c + d
    
```

```python
res = adds(5, ) # We are the assiging the value of b so it take the previous default one.
```

```python
print(res)  # So 5 + 10 becomes 15.
```

```python
res1 = adds(8, 4)  # Here a is the positional argument when we assign the value to default one it overwrite it.
```

```python
print(res1) # Here when we are giving the value it gives the 12.
```
