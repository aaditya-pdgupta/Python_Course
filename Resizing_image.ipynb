#### Image Resizing 
##### Image resizing is the process of changing the width and height of an image using the cv2.resize() function in OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Step 1: Read grayscale medical image
img = cv2.imread("X_ray.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Step 2:  Get original dimes=nsions
h, w = img.shape
print("Original Size", w, "x", h)
```

```python
#Step 3: Maintain aspect ratio
new_width = 256 # If we want the width remmain same we can
ratio = new_width / w
new_height = int(h * ratio)
```

```python
# Step 4: Resize using best method for shrinking
resized = cv2.resize(img, (new_width, new_height), interpolation = cv2.INTER_AREA)
```

```python
print("Resized Size:", new_width, "x", new_height)
```
